# [프로젝트명 : 정부의 정책으로 인한 전기차 및 충전소 보급 추이의 비교]

### 프로젝트 개요
- **기간:** 2024.08.12 ~ 2024.08.22 (총 7 영업일)
- **리더:** 최상근
- **스탭:** 김정훈, 한지우
- **정의:** 전기차 보급 목표 달성을 위한 개선 방안을 도출하고, 충전소 인프라의 균형적 분포를 위한 정책적 방향을 제시
- **목적:**
  1. 정부의 전기차 보급 및 충전소 설치 관련 정책이 실제로 전기차 보급과 충전 인프라 확충에 어떤 영향을 미쳤는지 분석합니다.
  2. 전기차와 충전소 보급 간의 불균형 문제를 식별하고, 이를 해결하기 위한 정책적 제안을 제시합니다.
  3. 향후 전기차 보급 목표 달성을 위한 정책적 지원 방향을 모색합니다.
- **범위:**
  1. 분석 대상: 2019년부터 2023년까지의 전기차 보급 현황, 충전소 설치 현황, 관련 정부 보조금 데이터.
  2. 분석 항목: 연도별 전기차 및 충전소 보급 추이, 보조금 지급 추이, 수도권과 비수도권의 충전소 설치 비율, 충전 인프라의 지역적 불균형 등.
  3. 데이터 출처: 국토교통부, 차지인포, 무공해차 통합누리집, 한국자동차모빌리티산업협회 등.
- **결론 및 향후 과제:**
  1. 정부의 친환경차 보급을 위한 예산은 지속적으로 증가했으나, 전기차와 충전소의 증가율은 둔화되는 경향을 보였습니다.
  2. 현재 추세로는 정부의 2030년 전기차 보급 목표(362만 대)를 달성하기 어려울 수 있으며, 충전소 인프라의 부족 및 지역 간 불균형이 주요 문제로 나타났습니다.
  3. 도시와 농촌 지역 간 충전소 분포의 불균형을 해소하기 위한 연구가 필요합니다.
  4. 전기차 사용자들의 행태를 분석하여 충전소 설치와 보조금 지원의 효율성을 높이는 방안을 모색해야 합니다.

### 일차별 계획

#### 1일차 (2024.08.12 - 월요일)
- **프로젝트 킥오프:**
  - 팀원 소개 및 역할 분담 확정 (리더: 최상근, 스탭: 김정훈, 한지우).
  - 프로젝트 목표 및 범위 재확인.
  - 데이터 수집 방법 및 도구 확정.
- **데이터 수집 준비:**
  - 공공 데이터 웹사이트의 API 접근 방법 및 관련 문서 리뷰.
  - 필요한 데이터 리스트업 (전기차 등록 정보, 충전소 설치 현황 등).
  - API 호출 테스트 및 초기 데이터 수집 시도.

#### 2일차 (2024.08.13 - 화요일)
- **데이터 수집 및 전처리:**

  - 수집된 데이터의 초기 전처리 (불필요한 데이터 제거, 결측치 처리).
  - 데이터 병합 및 정제 작업.
  - 수도권과 비수도권에 따른 데이터 구분 작업.

#### 3일차 (2024.08.14 - 수요일)
- **데이터 분석 1:**
  - 연도별 전기차 보급 및 충전소 설치 현황 분석.

  - 수도권과 비수도권 간의 전기차 및 충전소 설치 비율 비교.

#### 4일차 (2024.08.19 - 월요일)

  - 충전소를 이용할 수 있는 차량의 비율 분석.

#### 5일차 (2024.08.20 - 화요일)

  - 전기차 보급 목표 달성의 어려움 및 충전소 인프라의 불균형 문제 식별.
  - 향후 연구 방향 및 추가 과제 정의 (추가 데이터 수집, 사용자 행태 분석 등).
