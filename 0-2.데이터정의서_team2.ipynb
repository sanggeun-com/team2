# 데이터 정의서

## 1. 데이터 정의

### 1) `처리된_충전소(급완)분리.csv`
- **데이터 정의:** 이 데이터셋은 한국 내 전기차 충전소의 설치 정보 및 운영 현황을 나타냅니다. 각 충전소의 설치일자, 설치된 충전기의 종류, 수량, 설치 지역, 운영 기관 등의 정보를 포함합니다.

### 2) `2019_2023_수도권_비수도권_보조금평균.csv`
- **데이터 정의:** 이 데이터셋은 2019년부터 2023년까지 수도권과 비수도권에서 지급된 전기차 보조금의 평균 금액을 연도별로 기록한 데이터입니다.

### 3) `처리된_전국차량현황.csv`
- **데이터 정의:** 이 데이터셋은 한국 내 각 지역에서 등록된 차량의 종류 및 대수를 연도별로 기록한 데이터입니다. 특히 전기차 및 하이브리드 차량 등의 등록 현황을 포함하고 있습니다.

### 4) `11~24년 무공해차보조금.xlsx`
- **데이터 정의:** 이 데이터셋은 2011년부터 2024년까지 지급된 무공해차(전기차, 수소차 등) 보조금의 총액 및 지급 건수를 연도별로 기록한 데이터입니다.

## 2. 수집 방법

### 1) `처리된_충전소(급완)분리.csv`
- **수집 방법:** "차지인포"에서 제공된 데이터 통해 csv파일을 다운로드 하여 전기차 충전소 설치 현황 데이터를 수집하고, 이후 필요에 따라 전처리 및 분류 작업을 수행했습니다.

### 2) `2019_2023_수도권_비수도권_보조금평균.csv`
- **수집 방법:** "무공해차 통합누리집"에서 제공한 전기차 보조금 관련 데이터를 연도별로 집계된 파일을 다운로드하고, 수도권과 비수도권으로 구분하여 정리하여 csv파일로 저장했습니다.

### 3) `처리된_전국차량현황.csv`
- **수집 방법:** 국토교통부의 자동차 등록 데이터를 바탕으로 연도별 차량 등록 현황을 수집하고, 이후 데이터 전처리 및 분류 작업을 수행했습니다.

### 4) `11~24년 무공해차보조금.xlsx`
- **수집 방법:** 무공해차 통합누리집 및 환경부에서 제공하는 공공 데이터 자료를 기반으로 연도별 보조금 지급 데이터를 수집하고, 엑셀 파일로 정리했습니다.

## 3. 수집 경로

### 1) `처리된_충전소(급완)분리.csv`
- **수집 경로:** 차지인포에서 csv 파일로 저장

### 2) `2019_2023_수도권_비수도권_보조금평균.csv`
- **수집 경로:** 무공해차 통합누리집에서 csv 파일로 저장

### 3) `처리된_전국차량현황.csv`
- **수집 경로:** 국토교통부에서 수집한 데이터를 기반으로 구성.

### 4) `11~24년 무공해차보조금.xlsx`
- **수집 경로:** 무공해차 통합누리집에서 Excel  파일로 저장.

## 4. 데이터의 논리적 구조

### 1) `처리된_충전소(급완)분리.csv`
- **논리적 구조:**
  - `충전소_ID`: Primary Key
  - `설치일자`, `충전기_종류`, `설치지역`, `충전기_수량`, `운영기관`: Attributes

### 2) `2019_2023_수도권_비수도권_보조금평균.csv`
- **논리적 구조:**
  - `연도`: Primary Key
  - `수도권_평균_보조금`, `비수도권_평균_보조금`: Attributes

### 3) `처리된_전국차량현황.csv`
- **논리적 구조:**
  - `지역`, `연도`: Composite Primary Key
  - `차량종류`, `등록대수`: Attributes

### 4) `11~24년 무공해차보조금.xlsx`
- **논리적 구조:**
  - `연도`: Primary Key
  - `보조금_종류`, `지급액_총액`, `지급건수`: Attributes

## 5. 데이터의 물리적 저장 구조

### 1) `처리된_충전소(급완)분리.csv`
- **물리적 저장 구조:** CSV 파일 형식으로 저장. 각 필드는 쉼표(,)로 구분되며, 텍스트 데이터는 따옴표로 묶입니다.

### 2) `2019_2023_수도권_비수도권_보조금평균.csv`
- **물리적 저장 구조:** CSV 파일 형식으로 저장. 연도별 데이터가 행 단위로 기록되며, 필드 구분은 쉼표(,)를 사용합니다.

### 3) `처리된_전국차량현황.csv`
- **물리적 저장 구조:** CSV 파일 형식으로 저장. 각 행은 지역별, 연도별 차량 등록 현황을 나타내며, 필드 구분은 쉼표(,)로 처리됩니다.

### 4) `11~24년 무공해차보조금.xlsx`
- **물리적 저장 구조:** Excel 파일 형식으로 저장. 각 시트는 연도별 데이터를 포함하며, 필드 구분은 열 단위로 구성됩니다.
